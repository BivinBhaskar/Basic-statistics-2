```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Question 1

```python
df=pd.read_csv('F:\Excelr Assignments\Basic Statistics_Level_2\Table 1.txt',sep='\t')
```

```python
q1=pd.DataFrame(df)
```

```python
q1.head()
```

```python
q1['Measure X'] = pd.to_numeric(q1['Measure X'].str.replace('%', ''), errors='coerce')
```

```python
q1.head()
```

```python
plt.figure(figsize=(5,5))
plt.boxplot(q1['Measure X'])
plt.xlabel('Data')
plt.ylabel('Values')
plt.show()
```

```python
Q1 = np.quantile(q1['Measure X'],0.25)
Q3 = np.quantile(q1['Measure X'],0.75)
IQR = Q3 - Q1
upper_bound = Q3+(1.5*IQR)
lower_bound = Q1-(1.5*IQR)
Outliers=[q1[(q1['Measure X'] <= lower_bound) | (q1['Measure X']>= upper_bound)]]
Outliers
```

```python
plt.figure(figsize=(15,7))
a=sns.barplot(y=q1['Measure X'],x=q1['Name of company'])
for i in a.containers:
    a.bar_label(i,)
plt.xticks(rotation = 90, fontsize = 12, fontweight = 'bold')
plt.yticks(np.arange(0,100,10))
plt.show()
    
```
